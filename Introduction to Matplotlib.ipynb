# Matplotlib example workflow

```python
# 0. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')    #set plot style

# 1. prepare data
x=[1,2,3,4]
y = [11,22,33,44]

# 2. setup plot
fig,ax = plt.subplots(figsize = (10,10))    # figsize is width(10) and height(10) of the figure

# 3. Plot data
ax.plot(x,y);

# 4. Customise plot
ax.set(title = "Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save &show
fig.savefig("Images/sample-plot.png")
```

# Two options for subplots

```python
#Subplot option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                          figsize=(10,10))

#Plot to each different axis
ax1.plot(x,x*2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_price.keys(), nut_butter_price.values());
ax4.hist(np.random.random(10));
```

```python
#Subplot option 2
fig,ax = plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,10))
#Plot to each differnt index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_price.keys(), nut_butter_price.values());
ax[1,1].hist(np.random.random(10));
```

## pyplot vs matplotlib OO method....

* When plotting something quickly ,it is ok to use pyplot method.
* When plotting something more advanced, use the OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c = 'target');
```

```python
# OO method mixed with pyplot method
fig,ax = plt.subplots(figsize= (10,6))
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c = 'target',
            ax= ax);

ax.set_xlim([35,80])
```

```python
# OO method from scratch
fig,ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(x= over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

#add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# Subplot of chol, age,thalach

fig, (ax0,ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize= (10,13),
                             sharex = True);
plt.style.use('ggplot')
#add data to ax0
scatter = ax0.scatter(x = over_50.age,
                     y = over_50.chol,
                     c = over_50.target
                     );

# Customize the plot
ax0.set(title="Heart Disease and Cholesterol Levels",
       
       ylabel="Cholesterol");

#add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(over_50["chol"].mean(),
           linestyle="--");

#ad data to ax1
scatter = ax1.scatter(x = over_50.age,
                     y = over_50.thalach,
                     c = over_50.target);

#customize the plot
ax1.set(title = "Heart Disease and thalach levels",
       xlabel = "Age",
       ylabel = "Maximum Heart Rate");

#add a legend
ax1.legend(*scatter.legend_elements(), title = "Target");

#add a horizontal line
ax1.axhline(over_50.thalach.mean(),
           linestyle = "--");

#add a title to the figure
fig.suptitle("Heart Disease Analysis",
            fontsize = 50,
            fontweight = "bold");
```
