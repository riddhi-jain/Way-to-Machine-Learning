**Gradient Descent** is an optimization algorithm that's used when training a machine learning model. It's based on a convex function and tweaks its parameters iteratively to minimize a given function to its local minimum.

We want best fit for our machine learning model which is derived by minimizing the error(cost).

gradient descent.png


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math 
```

To get the new coefficients close to actual ones, we can either crease the learning rate or increase the number of iterations.
gradient descent2.png

```python
x = np.array([1,2,3,4,5])   # y = 3x + 2
y = np.array([5,8,11,14,17])
```

```python
def gradient(x, y):
  a = 0
  b = 0
  rate = 0.001          # learning rate or steps
  plt.scatter(x, y, color='Red');

  n = len(x)            # number of points 

  # we want the closest cost to initial cost
  cost_initial = 0

  for i in range(1000):
    y_pred = a*x + b  

    cost = (1/n)*sum([value**2 for value in (y-y_pred)])      # mse (mean square error) 
    plt.plot(x, y_pred);

    a_drev = -(2/n)*sum(x*(y-y_pred))    # derivative of mse wrt a
    b_drev = -(2/n)*sum(y-y_pred)        # derivative of mse wrt b

    a = a - rate*a_drev                  # new value of a
    b = b - rate*b_drev                   # new value of b 
    
    if math.isclose(cost, cost_initial):
      break
    cost_initial = cost
    return a, b
```

```python
df = pd.read_csv("/salaryData.csv")
df
```

```python
x = np.array(df.YearsExperience)
y = np.array(df.Salary)

m, c = gradient(x, y)    # y = m*x + c
```
